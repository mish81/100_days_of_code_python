# Functions with inputs

```python
# Basic Function Syntax
def greet():
    print("Hello")
    print("How do you do?")
    print("Isn't the weather nice today?")
```

```python
greet()
```

```python
# Passing variable to Function
def greet(name):
    print(f"Hello! {name}")
    print(f"How do you do? {name}")
    print(f"Hey {name}, Isn't the weather nice today?")
```

```python
greet("Dave")
```

```python
# Passing more than one parameter to the function
def greet(name, location):
    print(f"Hello! {name}")
    print(f"How do you do? {name}")
    print(f"Hey {name}, Isn't the weather nice today in {location} today?")
```

```python
greet("Camela","London")
```

# Positional vs Keyword Arguments

```python
# Passing more than one parameter to the function
def greet(name = "Ramesh", location = "Chennai"):
    print(f"Hello! {name}")
    print(f"How do you do? {name}")
    print(f"Hey {name}, Isn't the weather nice today in {location} today?")
```

```python
greet()
```
